```python
# Visualisasi untuk Text
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Membaca data
df = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

# Menggabungkan semua teks menjadi satu string
all_text = ' '.join(df['Text'].dropna())

# Memecah teks menjadi kata-kata dan menghitung frekuensi
words = all_text.split()
word_counts = Counter(words)

# Mengambil 10 kata paling umum
common_words = word_counts.most_common(10)
words, counts = zip(*common_words)

# Membuat DataFrame untuk visualisasi
word_freq_df = pd.DataFrame({'Kata': words, 'Frekuensi': counts})

# Visualisasi frekuensi kata
plt.figure(figsize=(10, 6))
sns.barplot(x='Frekuensi', y='Kata', data=word_freq_df)
plt.title('10 Kata Paling Umum dalam Ulasan')
plt.xlabel('Frekuensi')
plt.ylabel('Kata')
plt.show()
```

Distribusi untuk 'Text' adalah distribusi kategori karena setiap ulasan adalah teks yang berbeda. Visualisasi yang tepat untuk kolom ini bisa berupa analisis panjang teks dengan histogram atau analisis frekuensi kata untuk memahami tren umum dalam ulasan yang diberikan oleh pengguna.

### Penyebaran Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

# Menampilkan 5 baris pertama
print(df.head())

# Menghitung ukuran penyebaran untuk variabel numerik
helpfulness_numerator_range = df['HelpfulnessNumerator'].max() - df['HelpfulnessNumerator'].min()
helpfulness_denominator_range = df['HelpfulnessDenominator'].max() - df['HelpfulnessDenominator'].min()
score_range = df['Score'].max() - df['Score'].min()

helpfulness_numerator_variance = df['HelpfulnessNumerator'].var()
helpfulness_denominator_variance = df['HelpfulnessDenominator'].var()
score_variance = df['Score'].var()

helpfulness_numerator_std = df['HelpfulnessNumerator'].std()
helpfulness_denominator_std = df['HelpfulnessDenominator'].std()
score_std = df['Score'].std()

# Menampilkan hasil ukuran penyebaran
print("\nUkuran Penyebaran:")
print(f"Rentang HelpfulnessNumerator: {helpfulness_numerator_range}")
print(f"Rentang HelpfulnessDenominator: {helpfulness_denominator_range}")
print(f"Rentang Score: {score_range}")

print(f"\nVarians HelpfulnessNumerator: {helpfulness_numerator_variance}")
print(f"Varians HelpfulnessDenominator: {helpfulness_denominator_variance}")
print(f"Varians Score: {score_variance}")

print(f"\nDeviasi Standar HelpfulnessNumerator: {helpfulness_numerator_std}")
print(f"Deviasi Standar HelpfulnessDenominator: {helpfulness_denominator_std}")
print(f"Deviasi Standar Score: {score_std}")

# Visualisasi penyebaran dengan boxplot
plt.figure(figsize=(15, 5))

# Boxplot untuk HelpfulnessNumerator
plt.subplot(1, 3, 1)
sns.boxplot(x=df['HelpfulnessNumerator'])
plt.title('Boxplot HelpfulnessNumerator')

# Boxplot untuk HelpfulnessDenominator
plt.subplot(1, 3, 2)
sns.boxplot(x=df['HelpfulnessDenominator'])
plt.title('Boxplot HelpfulnessDenominator')

# Boxplot untuk Score
plt.subplot(1, 3, 3)
sns.boxplot(x=df['Score'])
plt.title('Boxplot Score')

plt.tight_layout()
plt.show()
```

### 1. **Rentang (Range)**
- **Rentang HelpfulnessNumerator**: 866
- **Rentang HelpfulnessDenominator**: 923
- **Rentang Score**: 4

**Penjelasan**:
- Rentang adalah selisih antara nilai maksimum dan minimum dari suatu variabel.
- Dalam kasus ini, nilai `HelpfulnessNumerator` bervariasi dari 0 hingga 866, menunjukkan bahwa ada sejumlah besar variasi dalam berapa banyak pengguna yang menganggap suatu ulasan membantu.

**Penjelasan**:

### analisis skewness

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

# Menghitung skewness
skewness = {
    'HelpfulnessNumerator': df['HelpfulnessNumerator'].skew(),
    'HelpfulnessDenominator': df['HelpfulnessDenominator'].skew(),
    'Score': df['Score'].skew()
}

print("Skewness:")
for var, skew in skewness.items():
    print(f"{var}: {skew}")

# Visualisasi distribusi untuk lebih memahami skewness
plt.figure(figsize=(12, 6))

# Histogram HelpfulnessNumerator
plt.subplot(1, 3, 1)
sns.histplot(df['HelpfulnessNumerator'], bins=30, kde=True)
plt.title('Distribusi HelpfulnessNumerator')

# Histogram HelpfulnessDenominator
plt.subplot(1, 3, 2)
sns.histplot(df['HelpfulnessDenominator'], bins=30, kde=True)
plt.title('Distribusi HelpfulnessDenominator')

# Histogram Score
plt.subplot(1, 3, 3)
sns.histplot(df['Score'], bins=30, kde=True)
plt.title('Distribusi Score')

plt.tight_layout()
plt.show()
```

### 1. **HelpfulnessNumerator: 37.52**
- **Interpretasi**: Nilai skewness yang sangat tinggi dan positif (37.52) menunjukkan bahwa distribusi dari `HelpfulnessNumerator` sangat miring ke kanan. Ini berarti banyak pengguna memberikan nilai rendah (seperti 0 atau 1), tetapi ada beberapa pengguna yang memberikan nilai tinggi. Dengan kata lain, sebagian besar ulasan tidak dianggap sangat membantu, tetapi ada beberapa ulasan yang dianggap sangat membantu.
  
### 2. **HelpfulnessDenominator: 32.63**
- **Interpretasi**: Seperti `HelpfulnessNumerator`, nilai skewness untuk `HelpfulnessDenominator` juga tinggi dan positif (32.63). Ini menunjukkan bahwa jumlah total penilaian membantu juga memiliki distribusi yang miring ke kanan. Artinya, banyak ulasan memiliki nilai rendah untuk denominasi ini, tetapi ada beberapa yang memiliki nilai tinggi. Ini mengindikasikan bahwa sebagian besar ulasan tidak mendapatkan banyak penilaian membantu, tetapi ada beberapa ulasan yang mendapatkan banyak penilaian.

### 3. **Score: -1.46**
- **Interpretasi**: Nilai skewness untuk `Score` adalah -1.46, yang menunjukkan distribusi miring ke kiri. Ini berarti bahwa sebagian besar ulasan cenderung memiliki skor yang lebih tinggi, tetapi ada beberapa ulasan dengan skor rendah. Dalam konteks ini, bisa diasumsikan bahwa banyak ulasan positif (skor 4 atau 5), sementara ada beberapa ulasan yang mendapatkan skor lebih rendah (1, 2, atau 3).

### Kesimpulannya:
- **HelpfulnessNumerator** dan **HelpfulnessDenominator** menunjukkan bahwa banyak ulasan tidak dianggap sangat membantu, tetapi ada beberapa yang sangat membantu, sedangkan **Score** menunjukkan bahwa sebagian besar ulasan positif, dengan beberapa ulasan negatif.
- Distribusi yang miring bisa memberikan wawasan tentang bagaimana pengguna merespons produk atau layanan.
